# 3. Data Cleansing

**Data Cleansing** (pembersihan data) adalah proses mendeteksi dan memperbaiki data yang kotor, tidak lengkap, atau tidak konsisten.

> 🗑️ *"Garbage in, garbage out"* — Analisis hanya sebaik datanya.

### Masalah data yang umum:
- Missing values (nilai kosong / NaN)
- Duplicate rows (baris duplikat)
- Inconsistent format (format tidak konsisten)
- Outliers (nilai ekstrem)
- Wrong data types (tipe data salah)

```python
import pandas as pd
import numpy as np

# Simulasi data kotor
data_kotor = {
    'ID'         : [1, 2, 3, 4, 5, 5, 6, 7],
    'Nama'       : ['Andi', 'Budi', None, 'Dian', 'Eko', 'Eko', 'Fani', 'Gita'],
    'Usia'       : [25, 30, 28, None, 35, 35, 22, 999],
    'Gaji'       : [8000000, np.nan, 10000000, 7500000, 15000000, 15000000, 9000000, 11000000],
    'Kota'       : ['Jakarta', 'bandung', 'SURABAYA', 'jakarta', 'Medan', 'Medan', None, 'Bali'],
    'Tanggal'    : ['2024-01-15', '15/02/2024', '2024-03-20', '25-04-2024',
                    '2024-05-10', '2024-05-10', '2024-06-01', '2024-07-08']
}
```

```python
df = pd.DataFrame(data_kotor)
print('=== DATA KOTOR ===')
display(df)
print(f'\nShape: {df.shape}')
```

## 3.1 Mendeteksi Missing Values

```python
# Cek missing values
print('=== MISSING VALUES PER KOLOM ===')
print(df.isnull().sum())

print('\n=== PERSENTASE MISSING ===')
persen_missing = (df.isnull().sum() / len(df) * 100).round(2)
print(persen_missing)
```

## 3.2 Menangani Missing Values

```python
df_clean = df.copy()

# Opsi 1: Hapus baris dengan missing values
# df_clean = df_clean.dropna()

# Opsi 2: Isi dengan nilai tertentu
df_clean['Nama'] = df_clean['Nama'].fillna('Unknown')

# Opsi 3: Isi numerik dengan mean/median
median_usia = df_clean['Usia'].median()
df_clean['Usia'] = df_clean['Usia'].fillna(median_usia)

mean_gaji = df_clean['Gaji'].mean()
df_clean['Gaji'] = df_clean['Gaji'].fillna(mean_gaji)

# Isi string dengan modus
modus_kota = df_clean['Kota'].mode()[0]
df_clean['Kota'] = df_clean['Kota'].fillna(modus_kota)

print('Missing values setelah dibersihkan:')
print(df_clean.isnull().sum())
```

```python
df_clean['Kota'].mode()[0]
```

## 3.3 Menangani Data Duplikat

```python
# Deteksi duplikat
print('Jumlah baris duplikat:', df_clean.duplicated().sum())
print('\nBaris yang duplikat:')
print(df_clean[df_clean.duplicated(keep=False)])

# Hapus duplikat (keep='first' → simpan kemunculan pertama)
df_clean = df_clean.drop_duplicates(keep='first').reset_index(drop=True)
print(f'\nShape setelah hapus duplikat: {df_clean.shape}')
```

```python
# Deteksi duplikat
print('Jumlah baris duplikat:', df_clean.duplicated().sum())
print('\nBaris yang duplikat:')
print(df_clean[df_clean.duplicated(keep=False)])

# Hapus duplikat (keep='first' → simpan kemunculan pertama)
df_clean1 = df_clean.drop_duplicates(keep='first')
df_clean1
```

## 3.4 Menangani Outliers
Outlier adalah nilai yang sangat jauh dari nilai lainnya. Kita gunakan metode **IQR (Interquartile Range)**.

```python
# Deteksi outlier dengan metode IQR
def detect_outliers_iqr(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[kolom] < lower) | (df[kolom] > upper)]
    return outliers, lower, upper

outliers_usia, low, high = detect_outliers_iqr(df_clean, 'Usia')
print(f'Batas normal Usia: {low:.0f} – {high:.0f}')
print('Outlier pada kolom Usia:')
print(outliers_usia[['Nama', 'Usia']])

# Hapus outlier
df_clean = df_clean[(df_clean['Usia'] >= low) & (df_clean['Usia'] <= high)]
df_clean = df_clean.reset_index(drop=True)
print(f'\nShape setelah hapus outlier: {df_clean.shape}')
```

# 4. Data Formatting

**Data Formatting** adalah proses menyeragamkan format data agar konsisten dan siap dianalisis.

Meliputi:
- Standarisasi teks (uppercase/lowercase)
- Format tanggal
- Rename kolom
- Format angka / mata uang
- Kategorisasi data

## 4.1 Formatting Text

```python
df_fmt = df_clean.copy()

print('Kota sebelum format:', df_fmt['Kota'].tolist())

# Standarisasi huruf kapital (Title Case)
df_fmt['Kota'] = df_fmt['Kota'].str.title()
df_fmt['Nama'] = df_fmt['Nama'].str.title()

print('Kota setelah format :', df_fmt['Kota'].tolist())

# Strip whitespace
df_fmt['Nama'] = df_fmt['Nama'].str.strip()

# Replace nilai
df_fmt['Kota'] = df_fmt['Kota'].str.replace('Dki Jakarta', 'Jakarta', regex=False)

print('\n--- String Methods Berguna ---')
contoh = '  Hello, Data Analyst!  '
print('upper() :', contoh.strip().upper())
print('lower() :', contoh.strip().lower())
print('title() :', contoh.strip().title())
print('len()   :', len(contoh.strip()))
```

## 4.2 Formatting Tanggal (DateTime)

```python
# Konversi kolom Tanggal ke format datetime
print('Tanggal sebelum:', df_fmt['Tanggal'].tolist())

df_fmt['Tanggal'] = pd.to_datetime(df_fmt['Tanggal'], infer_datetime_format=True, errors='coerce')

print('Tanggal sesudah:', df_fmt['Tanggal'].tolist())
print('Tipe data      :', df_fmt['Tanggal'].dtype)

# Ekstrak komponen tanggal
df_fmt['Tahun']     = df_fmt['Tanggal'].dt.year
df_fmt['Bulan']     = df_fmt['Tanggal'].dt.month
df_fmt['Hari']      = df_fmt['Tanggal'].dt.day
df_fmt['Nama_Hari'] = df_fmt['Tanggal'].dt.day_name()

print('\n--- Komponen Tanggal ---')
print(df_fmt[['Nama', 'Tanggal', 'Tahun', 'Bulan', 'Hari', 'Nama_Hari']])
```

## 4.3 Formatting Tipe Data & Angka

```python
# Cek tipe data saat ini
print('=== Tipe Data Sebelum ===')
print(df_fmt.dtypes)

# Konversi tipe data
df_fmt['Usia'] = df_fmt['Usia'].astype(int)
df_fmt['Gaji'] = df_fmt['Gaji'].astype(float)

print('\n=== Tipe Data Sesudah ===')
print(df_fmt[['Usia', 'Gaji']].dtypes)
```

```python
# Format angka sebagai mata uang (untuk display)
def format_rupiah(angka):
    return f'Rp {angka:,.0f}'

df_fmt['Gaji_Format'] = df_fmt['Gaji'].apply(format_rupiah)
print(df_fmt[['Nama', 'Gaji', 'Gaji_Format']])
```

## 4.4 Kategorisasi Data (Binning)

```python
# Kategorisasi usia
bins_usia  = [0, 25, 30, 40, 100]
label_usia = ['Junior', 'Mid-level', 'Senior', 'Expert']

df_fmt['Kategori_Usia'] = pd.cut(df_fmt['Usia'], bins=bins_usia, labels=label_usia)

# Kategorisasi gaji
bins_gaji  = [0, 8000000, 12000000, 20000000]
label_gaji = ['Low', 'Medium', 'High']

df_fmt['Kategori_Gaji'] = pd.cut(df_fmt['Gaji'], bins=bins_gaji, labels=label_gaji)

print(df_fmt[['Nama', 'Usia', 'Kategori_Usia', 'Gaji', 'Kategori_Gaji']])
```

## 4.5 Rename Kolom & Reorder

```python
# Rename kolom
df_final = df_fmt.rename(columns={
    'ID'   : 'employee_id',
    'Nama' : 'full_name',
    'Usia' : 'age',
    'Gaji' : 'salary',
    'Kota' : 'city'
})

# Pilih kolom yang relevan
kolom_pilihan = ['employee_id', 'full_name', 'age', 'city', 'salary',
                 'Kategori_Usia', 'Kategori_Gaji', 'Tanggal']
df_final = df_final[kolom_pilihan]

print('=== DATASET BERSIH & TERFORMAT ===')
print(df_final)
```

| 8 | `total` & `total_sales` kosong | Hitung ulang dari kolom lain |

```python
df = df_raw.copy()

# --- 1. Hapus kolom tidak relevan ---
df = df.drop(columns=['total', 'total_sales', 'Unnamed: 15'])
print(f'Setelah drop kolom  : {df.shape}')

# --- 2. Hapus baris missing values ---
df = df.dropna(how='any').reset_index(drop=True)
print(f'Setelah drop NA     : {df.shape}')

# --- 3. Hapus duplikat ---
df = df.drop_duplicates(keep='first').reset_index(drop=True)
print(f'Setelah drop dup    : {df.shape}')

# --- 4. Hapus quantity tidak valid ---
df = df[df['quantity'] > 0].reset_index(drop=True)
print(f'Setelah hapus qty<=0: {df.shape}')
```

```python
# --- 5. Bersihkan product_name ---
df['product_name'] = df['product_name'].str.strip().str.title()

# --- 6. Standarisasi category ---
df['category'] = df['category'].str.strip().str.title()

# --- 7. Standarisasi channel ---
channel_map = {
    'Toko_Oren_Online'   : 'Online - Toko Oren',
    'hijau_toko_web'     : 'Online - Toko Hijau',
    'Offline'            : 'Offline',
    'Online - Toko Oren' : 'Online - Toko Oren',
    'Online - Toko Hijau': 'Online - Toko Hijau'
}
df['channel'] = df['channel'].map(channel_map)

# --- Standarisasi teks lainnya ---
df['customer_name']    = df['customer_name'].str.strip().str.title()
df['customer_address'] = df['customer_address'].str.strip().str.title()

print('Nilai unik category setelah cleaning:', df['category'].unique())
print('Nilai unik channel  setelah cleaning:', df['channel'].unique())
```

## Step 4: Data Formatting

```python
df = df[df['sales_date'] != '12/31/2025']
```

```python
# Format tanggal
df['sales_date'] = pd.to_datetime(df['sales_date'], format='%d/%m/%Y')

# Ekstrak komponen tanggal
df['month']      = df['sales_date'].dt.month
df['month_name'] = df['sales_date'].dt.strftime('%B')
df['quarter']    = df['sales_date'].dt.quarter

# Perbaiki tipe data numerik
df['price']        = df['price'].astype(float)
df['quantity']     = df['quantity'].astype(int)
df['discount']     = df['discount'].astype(float)
df['shipping_fee'] = df['shipping_fee'].astype(float)
df['rating']       = df['rating'].astype(int)

# Hitung ulang kolom total dan total_sales
df['total']       = df['price'] * df['quantity']
df['total_sales'] = df['total'] * (1 - df['discount']) + df['shipping_fee']

# Tambah kolom kategorisasi
df['rating_label'] = df['rating'].apply(
    lambda r: 'Positif' if r >= 4 else ('Netral' if r == 3 else 'Negatif')
)

print('=== PERBANDINGAN SEBELUM VS SESUDAH ===')
print(f'Shape awal   : {df_raw.shape}')
print(f'Shape bersih : {df.shape}')
print(f'Baris dihapus: {df_raw.shape[0] - df.shape[0]}')
print()
print('=== PREVIEW DATASET BERSIH ===')
df[['order_id','sales_date','product_name','category','channel',
    'price','quantity','discount','shipping_fee','total','total_sales',
    'rating','rating_label','status']].head()
```

## Step 5: Analisis — Jawab Business Questions

```python
# ============================================================
# Q1: Revenue & komposisi per kategori produk?
# ============================================================
rev_kat = df.groupby('category').agg(
    Total_Revenue    = ('total_sales', 'sum'),
    Jumlah_Transaksi = ('order_id', 'count'),
    Total_Qty        = ('quantity', 'sum'),
    Avg_Transaksi    = ('total_sales', 'mean')
).sort_values('Total_Revenue', ascending=False).reset_index()

rev_kat['Share_%']     = (rev_kat['Total_Revenue'] / rev_kat['Total_Revenue'].sum() * 100).round(1)
rev_kat['Revenue_Fmt'] = rev_kat['Total_Revenue'].apply(lambda x: f'Rp {x:,.0f}')

print('=== Q1: REVENUE PER KATEGORI ===')
print(rev_kat[['category','Revenue_Fmt','Jumlah_Transaksi','Total_Qty','Share_%']].to_string(index=False))
```

```python
# ============================================================
# Q2: Channel penjualan paling menguntungkan?
# ============================================================
rev_channel = df.groupby('channel').agg(
    Total_Revenue    = ('total_sales', 'sum'),
    Jumlah_Transaksi = ('order_id', 'count'),
    Avg_Rating       = ('rating', 'mean')
).sort_values('Total_Revenue', ascending=False).reset_index()

rev_channel['Revenue_Fmt'] = rev_channel['Total_Revenue'].apply(lambda x: f'Rp {x:,.0f}')
rev_channel['Share_%']     = (rev_channel['Total_Revenue'] / rev_channel['Total_Revenue'].sum() * 100).round(1)
rev_channel['Avg_Rating']  = rev_channel['Avg_Rating'].round(2)

print('=== Q2: PERFORMA PER CHANNEL ===')
print(rev_channel[['channel','Revenue_Fmt','Jumlah_Transaksi','Avg_Rating','Share_%']].to_string(index=False))
```

```python
# ============================================================
# Q3: Tren penjualan per bulan?
# ============================================================
tren = df.groupby(['month','month_name']).agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order  = ('order_id', 'count')
).reset_index().sort_values('month')

print('=== Q3: TREN PENJUALAN BULANAN ===')
max_rev = tren['Total_Revenue'].max()
for _, row in tren.iterrows():
    bar = '█' * int(row['Total_Revenue'] / max_rev * 35)
    print(f"{row['month_name']:10} | {bar:<35} Rp {row['Total_Revenue']:>15,.0f}  ({row['Jumlah_Order']:,} order)")
```

```python
# ============================================================
# Q4: Produk paling laris?
# ============================================================
top_produk = df.groupby('product_name').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order  = ('order_id', 'count'),
    Total_Qty     = ('quantity', 'sum'),
    Avg_Rating    = ('rating', 'mean')
).sort_values('Total_Revenue', ascending=False).head(10).reset_index()

top_produk['Revenue_Fmt'] = top_produk['Total_Revenue'].apply(lambda x: f'Rp {x:,.0f}')
top_produk['Avg_Rating']  = top_produk['Avg_Rating'].round(2)

print('=== Q4: TOP 10 PRODUK BERDASARKAN REVENUE ===')
print(top_produk[['product_name','Revenue_Fmt','Jumlah_Order','Total_Qty','Avg_Rating']].to_string(index=False))
```

```python
# ============================================================
# Q5: Kota dengan penjualan tertinggi?
# ============================================================
rev_kota = df.groupby('customer_address').agg(
    Total_Revenue = ('total_sales', 'sum'),
    Jumlah_Order  = ('order_id', 'count'),
    Avg_Order_Val = ('total_sales', 'mean')
).sort_values('Total_Revenue', ascending=False).reset_index()

print('=== Q5: REVENUE PER KOTA ===')
max_rev = rev_kota['Total_Revenue'].max()
for _, row in rev_kota.iterrows():
    bar = '█' * int(row['Total_Revenue'] / max_rev * 30)
    print(f"{row['customer_address']:12} | {bar:<30} Rp {row['Total_Revenue']:>15,.0f}  ({row['Jumlah_Order']:,} order)")

print(f"\n🏆 Kota teratas: {rev_kota.iloc[0]['customer_address']} (Rp {rev_kota.iloc[0]['Total_Revenue']:,.0f})")
```

```python
# ============================================================
# Q6: Distribusi status order & rating pelanggan?
# ============================================================
status_dist = df['status'].value_counts().reset_index()
status_dist.columns = ['Status', 'Jumlah']
status_dist['Persentase'] = (status_dist['Jumlah'] / status_dist['Jumlah'].sum() * 100).round(1)

print('=== Q6A: DISTRIBUSI STATUS ORDER ===')
print(status_dist.to_string(index=False))

print()
rating_dist = df['rating'].value_counts().sort_index()
print('=== Q6B: DISTRIBUSI RATING PELANGGAN ===')
for rating, count in rating_dist.items():
    bintang = '⭐' * int(rating)
    bar     = '█' * int(count / rating_dist.max() * 30)
    print(f'{bintang} {bar:<30} {count:,}')

print(f'\nRata-rata rating: {df["rating"].mean():.2f} / 5')
```

```python
# ============================================================
# Q7: Pelanggan paling aktif
# ============================================================
top_customer = df.groupby('customer_name').agg(
    Jumlah_Order  = ('order_id', 'count'),
    Total_Belanja = ('total_sales', 'sum'),
    Avg_Order     = ('total_sales', 'mean')
).sort_values(['Jumlah_Order','Total_Belanja'], ascending=False).head(10).reset_index()

top_customer['Belanja_Fmt'] = top_customer['Total_Belanja'].apply(lambda x: f'Rp {x:,.0f}')
top_customer['Avg_Order']   = top_customer['Avg_Order'].round(0)

print('=== Q7: TOP 10 PELANGGAN PALING AKTIF ===')
print(top_customer[['customer_name','Jumlah_Order','Belanja_Fmt','Avg_Order']].to_string(index=False))
```

```python
# ============================================================
# Q8: Kategori paling laris di tiap channel
# ============================================================

kat_channel = df.groupby(['channel','category']).agg(
    Total_Revenue=('total_sales','sum'),
    Jumlah_Order=('order_id','count'),
    Total_Qty=('quantity','sum')
).reset_index()

# pivot supaya mudah dibandingkan
pivot_kat_channel = kat_channel.pivot(index='channel',
                                      columns='category',
                                      values='Total_Revenue').fillna(0)

pivot_kat_channel_fmt = pivot_kat_channel.applymap(lambda x: f'Rp {x:,.0f}')

print('=== Q8: PERFORMA KATEGORI DI SETIAP CHANNEL ===')
print(pivot_kat_channel_fmt)
```

## Step 6: Visualisasi Dashboard

```python
fig, axes = plt.subplots(4, 2, figsize=(18, 24))
fig.suptitle('Dashboard Penjualan FMCG 2025', fontsize=18, fontweight='bold', y=1.01)

COLORS = ['#1976D2', '#388E3C', '#F57C00', '#7B1FA2', '#D32F2F',
          '#0288D1', '#455A64', '#5D4037', '#C2185B', '#303F9F']

# --- 1. Revenue per Kategori (Horizontal Bar) ---
ax1 = axes[0, 0]
cat_data = df.groupby('category')['total_sales'].sum().sort_values()
bars = ax1.barh(cat_data.index, cat_data.values / 1e6, color=COLORS[:len(cat_data)])
ax1.set_title('Revenue per Kategori (Juta Rp)', fontweight='bold')
ax1.set_xlabel('Revenue (Juta Rp)')
for bar, val in zip(bars, cat_data.values):
    ax1.text(bar.get_width() + cat_data.max()/1e6*0.01,
             bar.get_y() + bar.get_height()/2,
             f'Rp {val/1e6:.1f}M', va='center', fontsize=9)
ax1.set_xlim(0, cat_data.max()/1e6 * 1.25)
ax1.grid(axis='x', alpha=0.3)

# --- 2. Market Share Channel (Pie) ---
ax2 = axes[0, 1]
ch_data = df.groupby('channel')['total_sales'].sum()
ax2.pie(ch_data.values, labels=ch_data.index, autopct='%1.1f%%',
        colors=COLORS[:len(ch_data)], startangle=140,
        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
ax2.set_title('Market Share per Channel', fontweight='bold')

# --- 3. Tren Bulanan (Line) ---
ax3 = axes[1, 0]
tren_data = df.groupby('month')['total_sales'].sum()
ax3.plot(tren_data.index, tren_data.values / 1e6, marker='o',
         color=COLORS[0], linewidth=2.5, markersize=8,
         markerfacecolor='white', markeredgewidth=2.5)
ax3.fill_between(tren_data.index, tren_data.values / 1e6, alpha=0.1, color=COLORS[0])
ax3.set_title('Tren Revenue Bulanan (Juta Rp)', fontweight='bold')
ax3.set_xlabel('Bulan')
ax3.set_ylabel('Revenue (Juta Rp)')
ax3.set_xticks(tren_data.index)
ax3.grid(True, alpha=0.3)

# --- 4. Top 8 Produk (Bar) ---
ax4 = axes[1, 1]
top8 = df.groupby('product_name')['total_sales'].sum().sort_values(ascending=False).head(8)
ax4.bar(range(len(top8)), top8.values / 1e6, color=COLORS[:len(top8)])
ax4.set_title('Top 8 Produk Berdasarkan Revenue', fontweight='bold')
ax4.set_ylabel('Revenue (Juta Rp)')
ax4.set_xticks(range(len(top8)))
ax4.set_xticklabels(top8.index, rotation=30, ha='right', fontsize=9)
ax4.grid(axis='y', alpha=0.3)

# --- 5. Revenue per Kota (Bar) ---
ax5 = axes[2, 0]
kota_data = df.groupby('customer_address')['total_sales'].sum().sort_values(ascending=False)
ax5.bar(range(len(kota_data)), kota_data.values / 1e6, color=COLORS[:len(kota_data)])
ax5.set_title('Revenue per Kota (Juta Rp)', fontweight='bold')
ax5.set_ylabel('Revenue (Juta Rp)')
ax5.set_xticks(range(len(kota_data)))
ax5.set_xticklabels(kota_data.index, rotation=30, ha='right', fontsize=9)
ax5.grid(axis='y', alpha=0.3)

# --- 6. Distribusi Status Order (Pie) ---
ax6 = axes[2, 1]
status_data = df['status'].value_counts()
status_colors = {'Sukses': '#388E3C', 'Pending': '#F57C00',
                 'Returned': '#1976D2', 'Failed': '#D32F2F'}
colors_status = [status_colors.get(s, '#9E9E9E') for s in status_data.index]
ax6.pie(status_data.values, labels=status_data.index, autopct='%1.1f%%',
        colors=colors_status, startangle=90,
        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5})
ax6.set_title('Distribusi Status Order', fontweight='bold')

# --- 7. Top Customer (Revenue + Order) ---
ax7 = axes[3, 0]

top_customer = df.groupby('customer_name').agg(
    Total_Revenue=('total_sales','sum'),
    Jumlah_Order=('order_id','count')
).sort_values('Jumlah_Order', ascending=False).head(10)

# Bar → Revenue
ax7.bar(range(len(top_customer)),
        top_customer['Total_Revenue']/1e6,
        color=COLORS[:len(top_customer)],
        alpha=0.8)

ax7.set_ylabel('Revenue (Juta Rp)')
ax7.set_title('Top 10 Pelanggan: Revenue vs Jumlah Order', fontweight='bold')

ax7.set_xticks(range(len(top_customer)))
ax7.set_xticklabels(top_customer.index, rotation=30, ha='right')

# Line → Order
ax7b = ax7.twinx()
ax7b.plot(range(len(top_customer)),
          top_customer['Jumlah_Order'],
          color='black',
          marker='o',
          linewidth=2)

ax7b.set_ylabel('Jumlah Order')

ax7.grid(axis='y', alpha=0.3)

# --- 8. Performa Kategori di Setiap Channel ---
ax8 = axes[3, 1]

kat_channel = df.groupby(['channel','category'])['total_sales'].sum().unstack()

kat_channel.plot(
    kind='bar',
    stacked=True,
    ax=ax8,
    color=COLORS[:len(kat_channel.columns)]
)

ax8.set_title('Performa Kategori di Setiap Channel', fontweight='bold')
ax8.set_xlabel('Channel')
ax8.set_ylabel('Revenue')
ax8.set_xticklabels(kat_channel.index, rotation=0)
ax8.legend(title='Kategori')

ax8.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('dashboard_fmcg.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Dashboard berhasil dibuat!')
```
